```python
top_10countries = data['Team'].value_counts().sort_values(ascending = False).head(10)
top_10countries
```

```python
plt.figure(figsize =(16,8))
plt.title('Top 10 countries')
sns.barplot(x=top_10countries.index,y=top_10countries ,palette ='Set1')
```

```python
plt.figure(figsize =(16,8))
plt.title('Age distribution of the athletes')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(data.Age,bins=np.arange(10,80,2),color='yellow',edgecolor='red')

```

```python
gender_count = data['Sex'].value_counts()
gender_count
```

```python
plt.figure(figsize = (16,8))
plt.title("Gender distribution")
plt.pie(gender_count,labels=gender_count.index,startangle=90,shadow=True,autopct="%1.1f%%")
```

```python
medals = data['Medal'].value_counts()
medals
```

```python
plt.figure(figsize = (16,8))
plt.title("Type of medal distribution")
plt.pie(medals,labels=medals.index,startangle=90,shadow=True,autopct="%1.1f%%")
```

```python
WomenOlympics = data[(data['Sex']== 'F')& (data['Season']== 'Summer')]
```

```python
plt.figure(figsize = (16,8))
plt.title("Women Participitaion")
sns.countplot(x='Year',data=WomenOlympics,palette = 'Spectral')
```
